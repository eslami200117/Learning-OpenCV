Import requirements.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Open image.

```python
img = cv2.imread('opencv_logo.jpg',1)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Write image.

```python
img = cv2.imread('OpenCV_Logo.jpg',0)
cv2.imshow('image',img)
key=cv2.waitKey(0)
if key==ord('s'):
   cv2.imwrite("opencv_logo_GS.jpg", img)
cv2.destroyAllWindows()
```

Plot image.

```python
img = cv2.imread('OpenCV_Logo.jpg',0)
plt.imshow(img)
plt.show()
```
